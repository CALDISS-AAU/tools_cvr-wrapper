```python
import requests
import pandas as pd
import numpy as np

cvr_virk_url = 'http://distribution.virk.dk/cvr-permanent/virksomhed/_search'
cvr_prod_url = 'http://distribution.virk.dk/cvr-permanent/produktionsenhed/_search'

headers = {
    'Content-Type': 'application/json',
}
```

```python
auth_bruger = ''
auth_pass = ''
```

```python
query_body = {"_source": ["Vrvirksomhed.cvrNummer", "Vrvirksomhed.bibranche1.branchekode"],
        "query": {
            "bool": {
                "must": {
                    "exists": {
                        "field": "Vrvirksomhed.bibranche1.branchekode"
                    }
                }
            }
        },
         "from": 0,
         "size": 100 # Max 3000
        }

query = str(query_body).replace('\'', '\"')
```

```python
query_body = {"_source": ["Vrvirksomhed.cvrNummer", "Vrvirksomhed.navne.navn", "Vrvirksomhed.navne.periode.gyldigTil", "Vrvirksomhed.beliggenhedsadresse", "Vrvirksomhed.bibranche1.branchekode", "Vrvirksomhed.bibranche1.branchetekst", "Vrvirksomhed.aarsbeskaeftigelse.intervalKodeAntalAnsatte", "Vrvirksomhed.aarsbeskaeftigelse.aar"],
        "query": {
            "bool": {
                "must": [
                    { "match": {"Vrvirksomhed.bibranche1.branchekode": "013000"}}
                ]
            }
        },
         "from": 0,
         "size": 100 # Max 3000
        }

query = str(query_body).replace('\'', '\"')
```

```python
response = requests.post(cvr_virk_url, headers = headers, auth = (auth_bruger, auth_pass), data = query)
```

```python
data_response = dict(response.json())

data = pd.json_normalize(data_response['hits']['hits'])
```

```python
exclude = ['_index', '_type', '_id', '_score']
data = data.loc[:, ~data.columns.isin(exclude)]
```

```python
json_columns = ['_source.Vrvirksomhed.beliggenhedsadresse', '_source.Vrvirksomhed.navne', '_source.Vrvirksomhed.aarsbeskaeftigelse', '_source.Vrvirksomhed.bibranche1']

data_long = data.copy()
for column_name in json_columns:
    data_long = data_long.explode(column_name).reset_index(drop = True)
```

```python
column_mapping = {'_source.Vrvirksomhed.beliggenhedsadresse': 'beliggenhedsadresse',
                 '_source.Vrvirksomhed.navne': 'navne',
                 '_source.Vrvirksomhed.aarsbeskaeftigelse': 'aarsbeskaeftigelse',
                 '_source.Vrvirksomhed.bibranche1': 'bibranche1'}

data_long = data_long.rename(columns = column_mapping)
```

```python
data_long.loc[40,:]
```

```python
data_long = data_long.dropna()
```

```python

json_columns = ['beliggenhedsadresse', 'navne', 'aarsbeskaeftigelse', 'bibranche1']

for column_name in json_columns:
    column_data = pd.json_normalize(data_long[column_name]).add_prefix(column_name + "_")
    
    data_long = data_long.merge(column_data, left_index = True, right_index = True)

data_long = data_long.loc[:, ~data_long.columns.isin(json_columns)]
data_long = data_long.drop_duplicates().reset_index(drop = True)
data_long = data_long.fillna(np.nan)
```

```python
data_long["navne_periode_aar"] = data_long['navne_periode.gyldigTil'].str.slice(stop=4).astype(float)
data_long["aarsbeskaeftigelse_aar"] = data_long["aarsbeskaeftigelse_aar"].astype(float)
data_long["beliggenhedsadresse_gyldigtil_aar"] = data_long['beliggenhedsadresse_kommune.periode.gyldigTil'].str.slice(stop=4).astype(float)
data_long["beliggenhedsadresse_gyldigfra_aar"] = data_long['beliggenhedsadresse_kommune.periode.gyldigFra'].str.slice(stop=4).astype(float)
```

```python
data_long = data_long.loc[data_long["aarsbeskaeftigelse_aar"] < data_long["navne_periode_aar"], :]
data_long = data_long.loc[(data_long["aarsbeskaeftigelse_aar"] < data_long["beliggenhedsadresse_gyldigtil_aar"]) | (data_long["aarsbeskaeftigelse_aar"] >= data_long["beliggenhedsadresse_gyldigfra_aar"]), :]
```
